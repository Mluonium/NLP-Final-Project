```python
# loading data 
def read_iob2_file(path):
    """
    read in iob2 file
    
    :param path: path to read from
    :returns: list with sequences of words and labels for each sentence
    """
    data = []
    current_words = []
    current_tags = []

    for line in open(path, encoding='utf-8'):
        line = line.strip()

        if line:
            if line[0] == '#':
                continue # skip comments
            tok = line.split('\t')

            current_words.append(tok[1])
            current_tags.append(tok[2])
        else:
            if current_words:  # skip empty lines
                data.append((current_words, current_tags))
            current_words = []
            current_tags = []

    # check for last one
    if current_tags != []:
        data.append((current_words, current_tags))
    return data
```

```python
train_data = read_iob2_file("en_ewt-ud-train.iob2")
dev_data = read_iob2_file("en_ewt-ud-dev.iob2")
print(len(train_data))
print(len(dev_data))
print(max([len(x[0]) for x in train_data]))
```

```python
# loading data 
def read_iob2_file2(path):
    """
    read in iob2 file
    
    :param path: path to read from
    :returns: list with sequences of words and labels for each sentence
    """
    data = []
    current_words = ""
    current_tags = []

    for line in open(path, encoding='utf-8'):
        line = line.strip()

        if line:
            if line[0] == '#':
                continue # skip comments
            tok = line.split('\t')

            current_words= current_words + tok[1]
        else:
            if current_words:  # skip empty lines
                data.append(current_words)
            current_words = ""

    # check for last one
    if current_tags != "":
        data.append(current_words)
    return data
```

```python
# loading tokenizer and multilingual BERT model 
from transformers import BertTokenizer, BertModel
tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased')
model = BertModel.from_pretrained("bert-base-multilingual-cased")
```

```python
# prepare data for BERT model by passing the strings through the BertTokenizer 
def tokenize_data(data, tokenizer):
    tokenized_data = []
    for tokens, labels in data:
        sentence = ' '.join(tokens)  # Concatenate tokens into a single string
        tokenized_sentence = tokenizer(sentence, return_tensors='pt')
        tokenized_data.append(tokenized_sentence)
    return tokenized_data
encoded_train_data = tokenize_data(train_data, tokenizer)
print(len(encoded_train_data))
```

```python
# passing inputs to BERT model
print(len(encoded_train_data))
# Iterate over each tokenized input and pass it to the BERT model
outputs = []
for encoded_input in encoded_train_data:
    # Pass the tokenized input to the BERT model
    output = model(input_ids=encoded_input['input_ids'], attention_mask=encoded_input['attention_mask'])
    outputs.append(output)
```
